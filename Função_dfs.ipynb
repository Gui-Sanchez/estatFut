```python
def dia(delta, formato='%Y-%m-%d', m=False):
  fuso_horario_brasilia = pytz.timezone('America/Sao_Paulo')
  data_hora_atual_brasilia = datetime.now(fuso_horario_brasilia) + timedelta(delta)
  if m:
    return data_hora_atual_brasilia.month
  else:
    return data_hora_atual_brasilia.strftime(formato)

def mes(n):
  if n == '01':
    return 'Janeiro'
  elif n == '02':
    return 'Fevereiro'
  elif n == '03':
    return 'Março'
  elif n == '04':
    return 'Abril'
  elif n == '05':
    return 'Maio'
  elif n == '06':
    return 'Junho'
  elif n == '07':
    return 'Julho'
  elif n == '08':
    return 'Agosto'
  elif n == '09':
    return 'Setembro'
  elif n == '10':
    return 'Outubro'
  elif n == '11':
    return 'Novembro'
  elif n == '12':
    return 'Dezembro'

def drop_reset_index(df):
    df = df.dropna()
    df = df.reset_index(drop=True)
    df.index += 1
    return df

def dia_sem_hora(data):
  data_sem_hora = datetime(data.year, data.month, data.day)
  return data_sem_hora

def remover_hora(data_com_hora_str):

  data_com_hora = datetime.strptime(data_com_hora_str, "%d/%m/%Y")
  data_sem_hora = datetime(data_com_hora.year, data_com_hora.month, data_com_hora.day)
  return data_sem_hora

def colunas(df):
  try:
    df = df[['League','Date','Home','Away','HT_Goals_H','HT_Goals_A','FT_Goals_H','FT_Goals_A','FT_Odds_Over05','FT_Odds_Over15','FT_Odds_Over25']]
    df.columns = ['League','Date','Home','Away','Goals_H_HT','Goals_A_HT','Goals_H_FT','Goals_A_FT',
                'Odds_Over05_FT', 'Odds_Over15_FT','Odds_Over25_FT']
    # Ajustando o Índice do DataFrame
    df['Date'] = df.apply(lambda row: dia_sem_hora(row['Date']), axis=1)
    df = df.sort_values(by='Date', ascending=True)
    df.reset_index(inplace=True, drop=True)
    df.index = df.index.set_names(['Nº'])
    df = df.rename(index=lambda x: x + 1)
    return df
  except:
    return df
```

# Função de análise

```python
# @title Tratamento dos dfs

def tratamento_dfs(df):

  df['Total_Goals_FT'] = df.apply(lambda row: (row['Goals_H_FT'] + row['Goals_A_FT']), axis=1)

  df['Over05_FT'] = df.apply(lambda row: 1 if (row['Total_Goals_FT'] > 0) else 0, axis=1)
  df['Over15_FT'] = df.apply(lambda row: 1 if (row['Total_Goals_FT'] > 1) else 0, axis=1)
  df['Over25_FT'] = df.apply(lambda row: 1 if (row['Total_Goals_FT'] > 2) else 0, axis=1)
  df['BTTS'] = df.apply(lambda row: 1 if (row['Goals_H_FT'] > 0 and row['Goals_A_FT'] > 0) else 0, axis=1)

  df['Porc_Over05FT_Home'] = (df.groupby('Home')['Over05_FT'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True) * 100)
  df['Porc_Over05FT_Away'] = (df.groupby('Away')['Over05_FT'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True) * 100)

  df['Porc_Over15FT_Home'] = (df.groupby('Home')['Over15_FT'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True) * 100)
  df['Porc_Over15FT_Away'] = (df.groupby('Away')['Over15_FT'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True) * 100)

  df['Porc_Over25FT_Home'] = (df.groupby('Home')['Over25_FT'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True) * 100)
  df['Porc_Over25FT_Away'] = (df.groupby('Away')['Over25_FT'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True) * 100)

  df['Porc_BTTS_Home'] = (df.groupby('Home')['BTTS'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True) * 100)
  df['Porc_BTTS_Away'] = (df.groupby('Away')['BTTS'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True) * 100)

  df['Media_Gols_Feitos_Home'] = df.groupby('Home')['Goals_H_FT'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True)
  df['Media_Gols_Sofridos_Home'] = df.groupby('Home')['Goals_A_FT'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True)

  df['Media_Gols_Feitos_Away'] = df.groupby('Away')['Goals_A_FT'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True)
  df['Media_Gols_Sofridos_Away'] = df.groupby('Away')['Goals_H_FT'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True)

  df['DP_Gols_Feitos_Home'] = df.groupby('Home')['Goals_H_FT'].rolling(window=5, min_periods=1).std().reset_index(0,drop=True)
  df['DP_Gols_Sofridos_Home'] = df.groupby('Home')['Goals_A_FT'].rolling(window=5, min_periods=1).std().reset_index(0,drop=True)

  df['DP_Gols_Feitos_Away'] = df.groupby('Away')['Goals_A_FT'].rolling(window=5, min_periods=1).std().reset_index(0,drop=True)
  df['DP_Gols_Sofridos_Away'] = df.groupby('Away')['Goals_H_FT'].rolling(window=5, min_periods=1).std().reset_index(0,drop=True)

  df['CV_Gols_Feitos_Home'] = (df['DP_Gols_Feitos_Home'] / df['Media_Gols_Feitos_Home'])
  df['CV_Gols_Sofridos_Home'] = (df['DP_Gols_Sofridos_Home'] / df['Media_Gols_Sofridos_Home'])

  df['CV_Gols_Feitos_Away'] = (df['DP_Gols_Feitos_Away'] / df['Media_Gols_Feitos_Away'])
  df['CV_Gols_Sofridos_Away'] = (df['DP_Gols_Sofridos_Away'] / df['Media_Gols_Sofridos_Away'])


  df1 = df[['Home','Away', 'Odds_Over05_FT', 'Odds_Over15_FT', 'Odds_Over25_FT','Goals_H_FT','Goals_A_FT','Total_Goals_FT','Porc_Over05FT_Home','Porc_Over05FT_Away','Porc_Over15FT_Home','Porc_Over15FT_Away','Porc_Over25FT_Home','Porc_Over25FT_Away','Porc_BTTS_Home','Porc_BTTS_Away','Media_Gols_Feitos_Home','Media_Gols_Sofridos_Home','Media_Gols_Feitos_Away','Media_Gols_Sofridos_Away','CV_Gols_Feitos_Home','CV_Gols_Sofridos_Home','CV_Gols_Feitos_Away','CV_Gols_Sofridos_Away']]
  df1 = df1.dropna()

  df_H = df1[['Home','Away','Porc_Over05FT_Home','Porc_Over15FT_Home','Porc_Over25FT_Home','Porc_BTTS_Home','Media_Gols_Feitos_Home','CV_Gols_Feitos_Home', 'Odds_Over05_FT', 'Odds_Over15_FT','Odds_Over25_FT', 'Goals_H_FT', 'Total_Goals_FT']]
  df_A = df1[['Home','Away','Porc_Over05FT_Away','Porc_Over15FT_Away','Porc_Over25FT_Away','Porc_BTTS_Away','Media_Gols_Sofridos_Away','CV_Gols_Sofridos_Away', 'Odds_Over05_FT', 'Odds_Over15_FT','Odds_Over25_FT', 'Goals_A_FT', 'Total_Goals_FT']]

  return df_H, df_A
```

```python
# @title Função de previsão

def previsoes_Over_goals(df_prev, df_H, df_A, n, liga, data_luc, p, steak=1):

  for i, j in zip(df_prev.Home, df_prev.Away):
    home = i
    away = j
    arq_luc = open('Lucro_prevs_'+str(dia(n))+'.txt', 'a+')
    flt_H = (df_H.Home == home)
    df10 = df_H[flt_H]
    df10 = df10.tail(1)
    flt_A = (df_A.Away == away)
    df20 = df_A[flt_A]
    df20 = df20.tail(1)
    jogo = ''

    arq_jogos = open('Jogos_Enviados.txt', 'r')
    lista_jogos_env = arq_jogos.readlines()
    arq_jogos.close()
    arq_jogos = open('Jogos_Enviados.txt', 'a+')

    try:

      if ((df10['Porc_Over05FT_Home'].item() >= 50) and
        (df10['Porc_BTTS_Home'].item() >= 20) and
        (df10['Media_Gols_Feitos_Home'].item() >= 1.0) and
        (df10['CV_Gols_Feitos_Home'].item() <= 1) and
        (df20['Porc_Over05FT_Away'].item() >= 50) and
        (df20['Porc_BTTS_Away'].item() >= 20) and
        (df20['Media_Gols_Sofridos_Away'].item() >= 1.0) and
        (df20['CV_Gols_Sofridos_Away'].item() <= 1) &
        (df20['Odds_Over05_FT'].item() >= 1.3)):

        if p == 'previsao':
          jogo = (f'{home} x {away} - 0,5 - {str(dia(n))}')
          l_aux = []
          if len(lista_jogos_env) > 0:
            for jogos in lista_jogos_env:
              if jogo in jogos:
                l_aux.append(1)
          if len(l_aux) > 0:
            mandou = 'sim'
          else:
            mandou = 'não'
          if mandou == 'sim':
            pass
          else:
            msg = (f'''
Liga:  {liga}
Data:  {str(dia(n, formato='%d/%m/%Y'))}
jogo:  {home} x {away}

Trabalhar o Over 0.5 FT
Odd: {df20["Odds_Over05_FT"].item()}

            ''')
            enviar_mensagem(msg)
            arq_jogos.write(jogo)
            arq_jogos.write('\n')
        else:
          r_h = df10['Goals_H_FT'].item()
          r_a = df20['Goals_A_FT'].item()

          if df20['Total_Goals_FT'].item() > 0.5:
            r = 'Green'
            luc = steak*(df20["Odds_Over05_FT"].item()-1)
          else:
            r = 'Red'
            luc = -steak

          msg = (f'''
Liga:  {liga}
Data:  {str(dia(n, formato='%d/%m/%Y'))}
jogo:  {home} x {away}

Trabalhar o Over 0.5 FT
Odd: {df20["Odds_Over05_FT"].item()}

Resultado: {home} {r_h} x {r_a} {away} - {r}
Lucro: R$ {luc:.2f}

          ''')
          enviar_mensagem(msg)
          arq_luc.write(str(luc))
          arq_luc.write('\n')
    except:pass

    try:
      if ((df10['Porc_Over15FT_Home'].item() >= 50) &
       (df10['Porc_BTTS_Home'].item() >= 40) &
       (df10['Media_Gols_Feitos_Home'].item() >= 1.0) &
       (df10['CV_Gols_Feitos_Home'].item() <= 1) &
       (df20['Porc_Over15FT_Away'].item() >= 50) &
       (df20['Porc_BTTS_Away'].item() >= 40) &
       (df20['Media_Gols_Sofridos_Away'].item() >= 1.0) &
       (df20['CV_Gols_Sofridos_Away'].item() <= 1) &
       (df20['Odds_Over15_FT'].item() >= 1.5)):

        if p == 'previsao':
          jogo = (f'{home} x {away} - 1,5 - {str(dia(n))}')
          l_aux = []
          if len(lista_jogos_env) > 0:
            for jogos in lista_jogos_env:
              if jogo in jogos:
                l_aux.append(1)
          if len(l_aux) > 0:
            mandou = 'sim'
          else:
            mandou = 'não'
          if mandou == 'sim':
            pass
          else:
            msg = (f'''
Liga:  {liga}
Data:  {str(dia(n, formato='%d/%m/%Y'))}
jogo:  {home} x {away}

Trabalhar o Over 1.5 FT
Odd: {df20["Odds_Over15_FT"].item()}

            ''')
            enviar_mensagem(msg)
            arq_jogos.write(jogo)
            arq_jogos.write('\n')
        else:
          r_h = df10['Goals_H_FT'].item()
          r_a = df20['Goals_A_FT'].item()

          if df20['Total_Goals_FT'].item() > 1.5:
            r = 'Green'
            luc = steak*(df20["Odds_Over15_FT"].item()-1)
          else:
            r = 'Red'
            luc = -steak

          msg = (f'''
Liga:  {liga}
Data:  {str(dia(n, formato='%d/%m/%Y'))}
jogo:  {home} x {away}

Trabalhar o Over 1.5 FT
Odd: {df20["Odds_Over15_FT"].item()}

Resultado: {home} {r_h} x {r_a} {away} - {r}
Lucro: R$ {luc:.2f}

          ''')
          enviar_mensagem(msg)
          arq_luc.write(str(luc))
          arq_luc.write('\n')
    except:pass

    try:
      if ((df10['Porc_Over25FT_Home'].item() >= 50) &
       (df10['Porc_BTTS_Home'].item() >= 60) &
       (df10['Media_Gols_Feitos_Home'].item() >= 1.0) &
       (df10['CV_Gols_Feitos_Home'].item() <= 1) &
       (df20['Porc_Over25FT_Away'].item() >= 50) &
       (df20['Porc_BTTS_Away'].item() >= 60) &
       (df20['Media_Gols_Sofridos_Away'].item() >= 1.0) &
       (df20['CV_Gols_Sofridos_Away'].item() <= 1) &
       (df20['Odds_Over25_FT'].item() >= 1.9)):


        if p == 'previsao':
          jogo = (f'{home} x {away} - 2,5 - {str(dia(n))}')
          l_aux = []
          if len(lista_jogos_env) > 0:
            for jogos in lista_jogos_env:
              if jogo in jogos:
                l_aux.append(1)
          if len(l_aux) > 0:
            mandou = 'sim'
          else:
            mandou = 'não'
          if mandou == 'sim':
            pass
          else:
            msg = (f'''
Liga:  {liga}
Data:  {str(dia(n, formato='%d/%m/%Y'))}
jogo:  {home} x {away}

Trabalhar o Over 2.5 FT
Odd: {df20["Odds_Over25_FT"].item()}

            ''')
            enviar_mensagem(msg)
            arq_jogos.write(jogo)
            arq_jogos.write('\n')
        else:
          r_h = df10['Goals_H_FT'].item()
          r_a = df20['Goals_A_FT'].item()

          if df20['Total_Goals_FT'].item() > 2.5:
            r = 'Green'
            luc = steak*(df20["Odds_Over25_FT"].item()-1)
          else:
            r = 'Red'
            luc = -steak

          msg = (f'''
Liga:  {liga}
Data:  {str(dia(n, formato='%d/%m/%Y'))}
jogo:  {home} x {away}

Trabalhar o Over 2.5 FT
Odd: {df20["Odds_Over25_FT"].item()}

Resultado: {home} {r_h} x {r_a} {away} - {r}
Lucro: R$ {luc:.2f}

          ''')
          enviar_mensagem(msg)
          arq_luc.write(str(luc))
          arq_luc.write('\n')

    except: pass

    arq_luc.close()
    arq_jogos.close()

  return
```
